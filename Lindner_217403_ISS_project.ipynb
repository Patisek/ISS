### Definitions of the functions for signal loading and playback

The function *sigs_load()* loads audio signals from files based on a list of their labels.  
The input parameter of the function is *labs (list of str)*, which is a list of strings representing the names of audio files to be loaded.  
The function returns a tuple containing a list of loaded audio signals *signals (list)*, the number of signals *N (int)*, and the sampling frequency *Fs (int)*.

```python
def sigs_load(labs):
  signals = []            # initialize signal list
  N = len(labs)           # total number of all signals

  for car in labs:                          # iterate through labels
    filename = login + "/" + car + ".wav"   # make a file path
    s, Fs = sf.read(filename)               # read the audio file
    signals.append(s)                       # append signal to signals list

  return signals, N, Fs          # returns signals, count, and sampling rate
```

### Definition of functions for checking Stationarity and Ergodicity and plotting s tatistics
The function *stat_plus_erg_check()* checks whether a signal is stationary or ergodic. This is determined by comparing segmental and overall estimates of variances and means, which this function provides. The signal is first segmented into individual frames, and then the relevant statistics are calculated to determine ergodicity, specifically at least those means and variances.

The input parameters of the function are the test signal *signal (ndarray)*, the frame size *fr_size (int)*, and the distance between frames *step (int)*.

The function returns a tuple containing a list of mean values for each frame *means_fr (list)*, a list of variances for each frame *vars_fr (list)*, the mean of the entire signal *mean_whole (float)*, and the variance of the entire signal *var_whole (float)*.


```python
def stat_plus_erg_check(sig, fr_size=1024, step=512):

    mean_whole = np.mean(sig)                            # mean value of entire signal
    var_whole = np.var(sig)                         # variance value of entire signal

    # Compute frame statistics
    num_frs = (len(sig) - fr_size) // step + 1  # total number of frames
    means_fr = []                                      # list to store means frame-wise
    vars_fr = []                                  # list to store variances frame-wise

    for i in range(num_frs):                               # iterate through frames
        fr = sig[i * step : i * step + fr_size] # extract frame
        means_fr.append(np.mean(fr))                    # compute mean of the frame
        vars_fr.append(np.var(fr))                 # compute variance of the frame

    return means_fr, vars_fr, mean_whole, var_whole
```

The function *sig_stats_plt()* displays graphs of the mean and variance for individual frames as well as for the entire signal.  
The input parameters of the function are a list of frame means *means_fr (list)*, a list of frame variances *vars_fr (list)*, the overall mean value of the signal *mean_whole (float)*, the index for correctly placing the signal when plotting *row_idx (int)*, and an array of Axes objects from the matplotlib library, which is the basic element for plotting graphs using matplotlib *axs (ndarray)*.


```python
def sig_stats_plt(means_fr, vars_fr, mean_whole, var_whole, row_idx, axs):
    # Plot frame-wise means on the left subplot
    axs[row_idx, 0].plot(means_fr, label="Frame-wise means")     # plot the fame-wise means
    axs[row_idx, 0].axhline(y=mean_whole, color='r', linestyle='--', label= "Overall mean") # add overall means as reference
    axs[row_idx, 0].set_ylabel("Mean")                         # lable y-axis as Mean
    axs[row_idx, 0].set_title(f"Signal {row_idx + 1} - Frame means vs. Overall mean") # add title
    axs[row_idx, 0].legend()                                   # display legend

    # Plot frame-wise variances on the right subplot
    axs[row_idx, 1].plot(vars_fr, label="Frame variances")   # frame-wise variances
    axs[row_idx, 1].axhline(y=var_whole, color='r', linestyle='--', label="Overall variance") # add overall variances as reference
    axs[row_idx, 1].set_ylabel("Variance")                           # lable y-axis asvariance
    axs[row_idx, 1].set_title(f"Signal {row_idx + 1} - Frame variances vs. Overall variance") # add title
    axs[row_idx, 1].legend()                                         # display legend
```

### Definition of functions for calculating MFCCs, their distances, and creating heatmaps
The function *get_mfccs()* calculates the MFCC coefficients for a given signal. The function utilizes the librosa library for advanced signal analysis by calling the function librosa.feature.mfcc(). This function has many possible input parameters, with the ones I have chosen and used being:  
y - the signal from which we are calculating the MFCCs,  
sr - the sampling frequency of the audio signal,  
n_mfcc - the number of coefficients we want the function to return,  
lifter - the liftering coefficient, if greater than 0, applies liftering (cepstral analysis) to the MFCCs, which can emphasize higher-order signal,  
n_fft - the length of the FFT window, which determines the resolution of the frequency analysis,  
hop_length - the number of samples between frames (this sets the overlap between frames),  
win_length = the size of the window of function window(), if not specified, the default value is equal to n_fft,  
window = the window function to apply to each frame, which can be of type str, tuple, number, function, or array specifying the window  
examples:
>'hamming' - commonly used, reduces sidelobes in the frequency domain
          'hann' - same as hamming with slightly different tapering
          'blackman' - better sidelobe suppression but wider main lobe
          ('kaiser', beta) - flexible window that can be adjusted using the shape parameter beta
          'boxcar' - rectangular window
          'triang' - triangular window

htk = if set to True, uses the HTK (hidden markov toolkit) formula for the Mel scale instead of the default Slaney formula
more information on possible parameters can be found at https://librosa.org/doc/latest/generated/librosa.feature.mfcc.html

The input parameters of the function are an array of the input signal *signal (array)*, the sampling frequency of the signals *Fs (int)*, the number of MFCCs we want to calculate (13 is the standard choice for speech recognition, 20-40 for more detailed analysis, and for higher resolution of details even bigger number can be chosen).
The function returns an array of calculated coefficients *mfccs (ndarray)*.


```python
def get_mfccs(sig, Fs, n_mfcc=13):
    mfccs = librosa.feature.mfcc(y=sig, sr=Fs, n_mfcc=n_mfcc, n_fft=2048, hop_length=512, win_length=2048, lifter = 7, window='hamming') # calculates mfccs
    return mfccs
```

The function *calc_dists()* calculates the distances between the MFCC coefficients of reference and test signals using the Euclidean metric (chosen because it is easy to implement).  
The input parameters of the function are a list of reference signals *sigs_refist)*, a list of test signals *sigs_tst (list)*, and the sampling frequency *sr (int)*.  
The function returns an array of distances between the reference and test signals *dists (ndarray)*.


```python
def calc_dists(sigs_ref, sigs_tst, sr):
    ref_mfccs = [get_mfccs(sig, sr) for sig in sigs_ref]     # compute mfccs for refernece signals
    tst_mfccs = [get_mfccs(sig, sr) for sig in sigs_tst]   # compute MFCCs for test signals

    dists = np.zeros((len(sigs_ref), len(sigs_tst)))             # make 2D distance array

    for i, ref_mfcc in enumerate(ref_mfccs):                                # loop through reference MFCCs
        for j, tst_mfcc in enumerate(tst_mfccs):                          # through test MFCCs
            dists[i, j] = np.linalg.norm(ref_mfcc - tst_mfcc)          # compute Euclidian distance

    return dists
```

The function *heatmap_plt()* displays a heatmap of distances between reference and test signals along with their values.
The input parameters of the function are an array of distances between reference and test signals *dists (ndarray)*, a list of test signals *sigs_tst (list)*, and a list of reference signals *sigs_ref (list)*.


```python
def heatmap_plt(dists, sigs_tst, sigs_ref):
    plt.figure(figsize=(12, 10))                                                  # new figure with given size
    img = plt.imshow(dists, cmap='viridis', origin='lower', aspect='auto')     # plot heatmap
    plt.colorbar(img, label="Euclidean distance")                                  # add colorbar with label
    plt.title("MFCC distances of reference and test signals")                 # add title to heatmap
    plt.xlabel("Test signal index")                                               # x-axis lable
    plt.ylabel("Reference signal index")                                          # y-axis lable
    plt.xticks(np.arange(len(sigs_tst)), [f"Test {i + 1}" for i in range(len(sigs_tst))])   # x-ticks
    plt.yticks(np.arange(len(sigs_ref)), [f"Ref {i + 1}" for i in range(len(sigs_ref))])      # y-ticks

    # distances betweeen values for heat map
    for i in range(len(sigs_ref)):
        for j in range(len(sigs_tst)):
            plt.text(j, i, f"{dists[i, j]:.2f}", ha="center", va="center", color="w")

    plt.show()
```

### Stationarity and Ergodicity
In the following section, we try to determine whether the signal is stationary or ergodic (in lectures it was said that this is the right approach) and proceed as follows:

Division of the signal into segments (frames).
* Estimation of parameters or functions (at least the mean value *a* and variance *D*).
* Estimation for each frame.
* Estimation for the entire signal.
* If the estimates for each frame are consistent, the signal is stationary. (Stationarity means that the statistical properties of the process do not change over time.)
* If the estimates for each frame correspond to the overall (time) estimate, the signal is also ergodic. (Ergodicity means that the time means are equal to the ensemble means, which means that the process is stationary and estimates are representative of overall behavior. This means that Ergodicity is a stronger condition than Stationarity.)
* If the parameters can be estimated from a single realization, the random signal is considered Ergodic and then time estimates are made.


### Plotting the check of Stationarity and Ergodicity of signals
The following part of the code ensures the plotting of graphs of the mean values and variances of individual frames of the signal as well as the signals as a whole. From the graphs, we can then determine whether the signal is stationary or ergodic by visual inspection.

#### Signal asignment
This code uses the Hungarian algorithm to assign test signals to reference signals in the most optimal way. We transpose the distance matrix (dists.T) because the algorithm works by assigning rows to columns, and in our case, rows represent reference signals and columns represent test signals. [3]

The reason for checking if the assigned value is larger than row average is to avoid making assignments that are wrong. If the assigned value is larger, it skips that assignment because it likely represents a poor match.

```python
# Hungarian algorithm
row_idxs, col_idxs = linear_sum_assignment(dists.T)   # finds optimal assignment

# Printing the results with a check for the largest value in each row
print("Optimal assignments:")
for tst, ref in zip(row_idxs, col_idxs):
    assigned_value = dists.T[tst, ref]
    row_avg = np.mean(dists.T[tst])  # Average value in the row

    # Check condition
    if assigned_value < row_avg:
        print(f"Test {tst+1} ({labs_tst[tst]}) -> Reference {ref+1} ({labs_ref[ref]}): Assignment is OK (value is smaller than the row average)")
    else:
        print(f"Test {tst+1} ({labs_tst[tst]}) -> Reference {ref+1} ({labs_ref[ref]}): Assignment skipped (value is not smaller than the row average)")
```
